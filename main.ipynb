```python
from ultralytics import YOLO

# 2nd model training with offline and online augmentation through Roboflow and Ultralytics
model = YOLO("yolo11m.pt")

model.train(
    data=r"Drone shot.v2i.yolov11\data.yaml",

    epochs=50,
    imgsz=960,

    batch=-1,
    patience=15,

    workers=4,
    device=0,

    augment=True,

    # Mild augmentation only
    mosaic=0.2,
    mixup=0.0,
    copy_paste=0.0,

    # Gentle spatial augmentation
    degrees=3.0,
    translate=0.02,
    scale=0.15,
    shear=0.0,
    perspective=0.0,

    # Realistic flips only
    fliplr=0.5,
    flipud=0.0,

    # Mild lighting variation
    hsv_h=0.01,
    hsv_s=0.2,
    hsv_v=0.2,
)
```

```python
from ultralytics import YOLO
import pandas as pd

# =========================
# LOAD TRAINED MODELS
# =========================

model1 = YOLO("runs\\detect\\train-4\\weights\\best.pt")      # YOLO11s Offline
model2 = YOLO("runs\\detect\\train-2\\weights\\best.pt")     # YOLO11m Offline+Online
model3 = YOLO("runs\\detect\\train-7\\weights\\best.pt")     # YOLO11m Online Only
```

```python


# =========================
# VALIDATION
# =========================

result1 = model1.val()
result2 = model2.val()
result3 = model3.val()

# =========================
# CREATE COMPARISON TABLE
# =========================

comparison = pd.DataFrame({
    "Model": [
        "YOLO11s (Offline Aug)",
        "YOLO11m (Offline + Online Aug)",
        "YOLO11m (Online Aug)"
    ],

    "mAP50": [
        result1.box.map50,
        result2.box.map50,
        result3.box.map50
    ],

    "mAP50-95": [
        result1.box.map,
        result2.box.map,
        result3.box.map
    ],

    "Precision": [
        result1.box.mp,
        result2.box.mp,
        result3.box.mp
    ],

    "Recall": [
        result1.box.mr,
        result2.box.mr,
        result3.box.mr
    ]
})

# =========================
# PRINT RESULTS
# =========================

print("\nMODEL COMPARISON\n")
print(comparison)

# =========================
# SAVE CSV FILE
# =========================

comparison.to_csv("model_comparison.csv", index=False)

print("\nComparison file saved as model_comparison.csv")
```

```python
import cv2
from ultralytics import YOLO


# Input video
video_path = "Demo.mp4"
cap = cv2.VideoCapture(video_path)

# Output video 

output_path = f"output_{video_path}"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(
    output_path,
    fourcc,
    20,
    (
        int(cap.get(3)),
        int(cap.get(4))
    )
)

# Original dataset classes
CLASS_NAMES = {
    0: 'awning-tricycle',
    1: 'bicycle',
    2: 'bus',
    3: 'car',
    4: 'motor',
    5: 'pedestrian',
    6: 'people',
    7: 'tricycle',
    8: 'truck',
    9: 'van'
}

# Human classes
HUMAN_CLASSES = [5, 6]
# Car classes (car and van)
CAR_CLASSES = [3, 9]

while cap.isOpened():

    ret, frame = cap.read()

    if not ret:
        break

    results = model1(frame)

    human_count = 0
    car_count = 0

    for result in results:

        boxes = result.boxes

        for box in boxes:

            cls = int(box.cls[0])
            conf = float(box.conf[0])

            x1, y1, x2, y2 = map(int, box.xyxy[0])

            class_name = CLASS_NAMES.get(cls, 'unknown')

            # Only process human or car (car/van)
            if cls in HUMAN_CLASSES:
                human_count += 1

                label = f"{class_name}: {conf:.2f}"
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(
                    frame,
                    label,
                    (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.5,
                    (0, 255, 0),
                    2
                )

            elif cls in CAR_CLASSES:
                car_count += 1

                label = f"{class_name}: {conf:.2f}"
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
                cv2.putText(
                    frame,
                    label,
                    (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.5,
                    (255, 0, 0),
                    2
                )

    # Display counts on frame
    cv2.putText(
        frame,
        f"Humans: {human_count}",
        (50, 50),
        cv2.FONT_HERSHEY_SIMPLEX,
        1.5,
        (0, 255, 0),
        3
    )

    cv2.putText(
        frame,
        f"Cars: {car_count}",
        (50, 100),
        cv2.FONT_HERSHEY_SIMPLEX,
        1.5,
        (255, 0, 0),
        3
    )

    out.write(frame)

cap.release()
out.release()
print(f"✅ Processing complete. Saved to {output_path}")
```
